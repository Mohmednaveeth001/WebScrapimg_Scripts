```
import re


data = """
Bottom torso:                     70% Polyamide(Nylon), 30% Elastane(Spandex)
                
                
Inner:                     90% Polyester, 10% Elastane(Spandex)
                
                
Main:                     100% Polyester
                
        """



d4 = []
d1 = data.strip()
d2 = d1.split("\n")
d3 = [i.strip() for i in d2 if i.strip()]
for data in d3:
  temp_data = re.sub(r"\s+"," ",data)
  d4.append(temp_data)

output = ' '.join(d4)
print(output)
```
